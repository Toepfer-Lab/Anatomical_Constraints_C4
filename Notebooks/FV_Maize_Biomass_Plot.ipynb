# Maize Biomass Components

## Imports

```python
# importing the required library
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Getting the Dataset

```python
maize = pd.read_csv("..\Datasets\Maize_Leaf_Biomass.csv")

groupedvalues = maize.groupby('Classes').sum()

groupedvalues
```

## Plotting

```python
maize = maize.sort_values('Maize',ascending=False).reset_index()

plt.figure(figsize=(8, 4))

sns.set_style("ticks")

plot = sns.barplot(x = 'Classes',
            y = 'Maize',
            data = maize,
           ci = None, estimator=sum) 


plot.set_yticklabels(plot.get_yticks(), size = 10)


# set labels

plt.xlabel("Biomass Components", size=12)
plt.ylabel("Relative Percentage (%)", size=12)
plt.title("Maize Biomass Composition", size=14, fontweight="bold")
plt.tight_layout()
```
